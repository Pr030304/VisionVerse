```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from skimage.restoration import denoise_bilateral
images=['1.png','2.jpg']
```

**GAUSSIAN FILTER**

```python
for image in images:
  image = cv2.imread(image)
  kernel_size = 9
  sigma = 1.0
  blur = cv2.GaussianBlur(image, (kernel_size, kernel_size), sigmaX=sigma, sigmaY=sigma)
  cv2_imshow(blur)
```

**MEDIAN FILTER**

```python
for image in images:
  image = cv2.imread(image)
  kernel_size = 5
  median = cv2.medianBlur(image, kernel_size)
  cv2_imshow(median)
```

**LAPLACIAN FILTER**

```python
for image in images:
  image = cv2.imread(image)
  laplacian = cv2.Laplacian(image, cv2.CV_64F)
  cv2_imshow(laplacian)
```

**BILATERAL FILTER**

```python
for image in images:
  image = cv2.imread(image)
  filtered_image = cv2.bilateralFilter(image, 9, 75, 75)
  cv2_imshow(filtered_image)
```

**BOX FILTER**

```python
for image in images:
  image = cv2.imread(image)
  kernel_size = 5
  box = cv2.boxFilter(image, -1, (kernel_size, kernel_size))
  cv2_imshow(box)
```
