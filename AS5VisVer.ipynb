```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
from google.colab.patches import cv2_imshow
```

```python
images = [
    "1.png",
    "2.png",
    "3.png",
    "4.png",
    "5.png",
    "6.jpg",
]
K_values = [5, 25, 50]
segmented_images=[];
print("K-Means Clustered Images")
for image_path in images:
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    data = image.reshape((-1, 3))
    for K in K_values:
        kmeans = KMeans(n_clusters=K, n_init='auto',random_state=0).fit(data)
        segmented_image = kmeans.cluster_centers_[kmeans.labels_].reshape(image.shape)
        segmented_images.append(segmented_image)
        cv2_imshow(segmented_image)
```

```python
image = cv2.imread("boat.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh_otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
thresh_normal = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)[1]
print("Original Image")
cv2_imshow(image)
print("Otsu Thresholding")
cv2_imshow(thresh_otsu)
print("Normal Thresholding")
cv2_imshow(thresh_normal)
```

Otsu Thresholding yields better result.
